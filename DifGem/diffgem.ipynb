```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt


r = np.arange(0, 3.0, 0.01)
theta = np.pi * r

ax = plt.subplot(111, projection='polar')
ax.plot(theta, r, color='r', linewidth=3)
ax.set_rmax(2.0)
ax.grid(True)

ax.set_title("A line plot on a polar axis", va='bottom')
plt.show()
```

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt

r1 = 1 + np.cos(theta)
r2 = 1 - np.cos(theta)
theta = np.linspace(0, 2*np.pi, 100)  

ax = plt.subplot(111, polar=True)      # add subplot in polar coordinates 
#                  axisbg='Azure')       # background colourax.plot(theta, r, color='r', linewidth=3)
ax.set_rmax(2.0)
ax.grid(True)

ax.plot(theta, r1,
        color='purple',                # line colour
        ls='--',                       # line style
        lw=1,                          # line width
        label='p_1')        # label

ax.plot(theta, r2, 
        color='purple',
        linewidth=1,
        ls = '-',
        label = 'p_2')

ax.legend(loc="lower right")           # legend location
ax.set_xticklabels([])
ax.set_yticklabels([])

plt.show()
```

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp


x = np.arange(-2.0, 5.0, 0.001)
y1 = np.sin(x)
y2 = -1/2*((x-0.78)*(x-0.78))+np.pi*(x-0.78)/2+1-np.pi*np.pi/8+0.6

ax = plt.subplot(111)
ax.plot(x, y1, color='r', linewidth=1, ls = '--', label="sin(x)")
ax.plot(x, y2, color='g', linewidth=1, ls = '-', label="parabola")
ax.axhline()
ax.axvline()
ax.set_yticks(np.arange(-5,5,1))
ax.set_xticks(np.arange(-5,5,0.5))

ax.grid(True)

ax.legend(loc="lower right") 

plt.show()
```
